```python
from ML_models.prepare_object_for_prediction import encode_categorical_columns
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler



data = pd.read_csv(dataset_path, on_bad_lines='skip')
data = data.drop(data.columns[0], axis=1)
data = data.drop(['property_type'], axis=1)
data.dropna(inplace=True)
data = data.drop('address', axis=1)
data = data.drop(data.columns[0], axis=1)

categorical_columns = [
    'community'
]

data = encode_categorical_columns(data, categorical_columns)
data['price'] = data['price'].apply(lambda x: math.ceil(x / 500000) * 500000)


X = data.drop("price", axis=1)  # Remove price column to create input data
y = data["price"]  # Target variable is the price column


# Scale numerical features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```
